```python
#question 23
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('application_train.csv')

df['CREDIT_INCOME_RATIO'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']

df = df[df['CREDIT_INCOME_RATIO'] < 10]

bins = [0, 0.5, 1, 1.5, 2, 3, 5, 10]
labels = ['0–0.5', '0.5–1', '1–1.5', '1.5–2', '2–3', '3–5', '5–10']
df['CREDIT_INCOME_BIN'] = pd.cut(df['CREDIT_INCOME_RATIO'], bins=bins, labels=labels)

ratio_default_rate = df.groupby('CREDIT_INCOME_BIN')['TARGET'].mean() * 100
ratio_default_rate = ratio_default_rate.round(2)

print("Default Rate (%) by Credit to Income Ratio Group:")
print(ratio_default_rate)

plt.figure(figsize=(10, 5))
ratio_default_rate.plot(kind='bar', color='tomato', edgecolor='black')
plt.title("Default Rate vs. Credit to Income Ratio")
plt.xlabel("Credit to Income Ratio Group")
plt.ylabel("Default Rate (%)")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#question 25
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv("application_train.csv")

# Step 2: Replace 365243 (invalid employment) with NaN
df["DAYS_EMPLOYED"] = df["DAYS_EMPLOYED"].replace(365243, pd.NA)

# Step 3: Convert DAYS_EMPLOYED to positive years
df["EMPLOYMENT_YEARS"] = (-df["DAYS_EMPLOYED"]) / 365

# Step 4: Drop NaNs
df = df.dropna(subset=["EMPLOYMENT_YEARS"])

# Step 5: Create bins
bins = [0, 2, 5, 10, 20, 30, 50]
labels = ["0–2 yrs", "2–5 yrs", "5–10 yrs", "10–20 yrs", "20–30 yrs", "30–50 yrs"]
df["EMPLOYMENT_BIN"] = pd.cut(df["EMPLOYMENT_YEARS"], bins=bins, labels=labels)

# Step 6: Calculate default rate
default_rate = df.groupby("EMPLOYMENT_BIN")["TARGET"].mean() * 100
default_rate = default_rate.round(2)

# Step 7: Plot
plt.figure(figsize=(10, 5))
default_rate.plot(kind="bar", color="coral", edgecolor="black")
plt.title("Default Rate by Employment Duration")
plt.xlabel("Employment Duration Group")
plt.ylabel("Default Rate (%)")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
# question 29
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv('application_train.csv')

# Step 2: Plot each EXT_SOURCE variable
ext_cols = ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']

for col in ext_cols:
    plt.figure(figsize=(10, 5))
    sns.kdeplot(data=df[df['TARGET'] == 0], x=col, label='Repaid (TARGET=0)', fill=True, color='green')
    sns.kdeplot(data=df[df['TARGET'] == 1], x=col, label='Defaulted (TARGET=1)', fill=True, color='red')
    
    plt.title(f'Distribution of {col} by Default Status')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
#question 30
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load data
df = pd.read_csv('application_train.csv')

# Step 2: Clean and convert DAYS_BIRTH to Age
df['AGE'] = (-df['DAYS_BIRTH']) / 365  # in years

# Step 3: Remove NaNs in EXT_SOURCE_1
df_filtered = df[['AGE', 'EXT_SOURCE_1']].dropna()

# Step 4: Plot relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_filtered, x='AGE', y='EXT_SOURCE_1', alpha=0.3, color='blue', edgecolor=None)
sns.regplot(data=df_filtered, x='AGE', y='EXT_SOURCE_1', scatter=False, color='red', label='Trend Line')

plt.title('Relationship between Age and EXT_SOURCE_1')
plt.xlabel('Applicant Age (Years)')
plt.ylabel('EXT_SOURCE_1 Score')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#question 33
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv('application_train.csv')

# Step 2: Drop missing values in inquiry column
df_filtered = df.dropna(subset=['AMT_REQ_CREDIT_BUREAU_YEAR'])

# Step 3: Group by number of inquiries and compute default rate
inquiry_default_rate = df_filtered.groupby('AMT_REQ_CREDIT_BUREAU_YEAR')['TARGET'].mean() * 100
inquiry_default_rate = inquiry_default_rate.round(2)

# Step 4: Show results
print("Default Rate (%) by Number of Bureau Inquiries (Past Year):")
print(inquiry_default_rate)

# Step 5: Plot
plt.figure(figsize=(10, 5))
inquiry_default_rate.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Default Rate by Credit Bureau Inquiries (Last 1 Year)')
plt.xlabel('Number of Inquiries')
plt.ylabel('Default Rate (%)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#question 35
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('application_train.csv')

pivot = df.pivot_table(
    index='NAME_FAMILY_STATUS',
    columns='NAME_INCOME_TYPE',
    values='AMT_CREDIT',
    aggfunc='mean'
).round(0)

plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap='YlGnBu', linewidths=0.5)

plt.title('Average Loan Amount by Family Status and Income Type')
plt.xlabel('Income Type')
plt.ylabel('Family Status')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
