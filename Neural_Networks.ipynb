```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD

# Input data (AND gate)
X = np.random.randint(0,10,(300000,121))

# Output labels
y = np.random.randint(0,1,[300000])

# Build the model
model = Sequential()
model.add(Dense(5, input_dim=121, activation='relu'))   # Hidden layer with 5 nodes
model.add(Dense(1, activation='sigmoid'))             # Output layer for binary classification

# Compile the model
model.compile(optimizer=SGD(learning_rate=0.1),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the model and store the training history
history = model.fit(X, y, epochs=100, verbose=0)

# Evaluate the model
loss, accuracy = model.evaluate(X, y, verbose=0)
print(f"\nFinal Accuracy: {accuracy * 100:.2f}%")

# Predict
print("\nPredictions:")
print(model.predict(X))

# Plot loss and accuracy graphs
plt.figure(figsize=(12, 5))

# Plot Loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], color='red', label='Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plot Accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], color='blue', label='Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```

# 📘 Deep Learning Terminology

## 🔧 Basic Terms

- **Neuron**: Basic unit of a neural network that receives input, applies a weight and bias, and passes it through an activation function.
- **Layer**: A collection of neurons. Types include Input, Hidden, and Output layers.
- **Activation Function**: Function that introduces non-linearity (e.g., `relu`, `sigmoid`, `tanh`, `softmax`).
- **Weight**: A trainable parameter representing the strength of a connection between neurons.
- **Bias**: A trainable parameter added to the input of the activation function to shift it.
- **Forward Propagation**: The process of passing inputs through the network to get the output.
- **Loss Function**: Measures the difference between predicted and actual output (e.g., `MSE`, `binary_crossentropy`).
- **Backpropagation**: Algorithm for updating weights by propagating the error backward through the network.
- **Epoch**: One complete pass through the entire training dataset.
- **Batch Size**: Number of samples processed before the model updates.
- **Learning Rate**: Controls how much to update weights during training.
- **Optimizer**: Algorithm that updates model parameters (e.g., `SGD`, `Adam`, `RMSprop`).

## 🏗️ Network Types

- **Perceptron**: The simplest neural network with a single neuron.
- **Feedforward Neural Network (FNN)**: Network where connections do not form cycles.
- **Convolutional Neural Network (CNN)**: Used for image data; uses convolution layers.
- **Recurrent Neural Network (RNN)**: Suitable for sequential data like text and time series.
- **Autoencoder**: Unsupervised model that learns to compress and reconstruct input data.
- **GAN (Generative Adversarial Network)**: Two networks (Generator & Discriminator) trained in competition.

## 🧪 Evaluation Metrics

- **Accuracy**: `(Correct Predictions / Total Predictions)`
- **Precision**: `TP / (TP + FP)`
- **Recall**: `TP / (TP + FN)`
- **F1 Score**: `2 * (Precision * Recall) / (Precision + Recall)`
- **Confusion Matrix**: Table showing true vs predicted classifications.

## 📊 Regularization & Techniques

- **Overfitting**: When a model performs well on training data but poorly on unseen data.
- **Underfitting**: When a model performs poorly on both training and test data.
- **Dropout**: Randomly disables neurons during training to prevent overfitting.
- **Batch Normalization**: Normalizes inputs of each layer for faster training.
- **Early Stopping**: Stop training when validation performance starts degrading.

## 🗃️ Data Terms

- **Training Set**: Data used to train the model.
- **Validation Set**: Data used to tune the model.
- **Test Set**: Data used to evaluate final model performance.
- **One-hot Encoding**: Converts categorical variables to binary vector format.

## ⚙️ Others

- **Transfer Learning**: Using a pretrained model on a new problem.
- **Fine-Tuning**: Training only a few layers of a pretrained model.
- **Hyperparameters**: Settings like learning rate, batch size, etc., not learned from data.
- **Model Parameters**: The weights and biases learned by the model.


```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from keras import backend as K

# Define 20 activation functions and their formulas
activation_functions = {
    "linear": lambda x: x,
    "sigmoid": tf.keras.activations.sigmoid,
    "tanh": tf.keras.activations.tanh,
    "relu": tf.keras.activations.relu,
    "softplus": tf.keras.activations.softplus,
    "softsign": tf.keras.activations.softsign,
    "selu": tf.keras.activations.selu,
    "elu": tf.keras.activations.elu,
    "exponential": tf.keras.activations.exponential,
    "hard_sigmoid": tf.keras.activations.hard_sigmoid,
    "swish": lambda x: x * tf.keras.activations.sigmoid(x),
    "mish": lambda x: x * tf.math.tanh(tf.math.softplus(x)),
    "leaky_relu": lambda x: tf.nn.leaky_relu(x, alpha=0.2),
    "prelu": lambda x: tf.maximum(0.1 * x, x),
    "gelu": tf.keras.activations.gelu,
    "bent_identity": lambda x: ((np.sqrt(x ** 2 + 1) - 1) / 2) + x,
    "arctangent": lambda x: tf.math.atan(x),
    "sinc": lambda x: tf.where(x == 0, tf.ones_like(x), tf.math.sin(x) / x),
    "gaussian": lambda x: tf.exp(-tf.square(x)),
    "sign": lambda x: tf.sign(x),
}

# Input range for plotting
x_vals = np.linspace(-10, 10, 500)
x_tensor = tf.convert_to_tensor(x_vals, dtype=tf.float32)

# Plotting
plt.figure(figsize=(20, 20))
for i, (name, func) in enumerate(activation_functions.items()):
    plt.subplot(5, 4, i+1)
    y_vals = func(x_tensor)
    plt.plot(x_vals, y_vals.numpy())
    plt.title(name)
    plt.grid(True)

plt.tight_layout()
plt.suptitle("Activation Functions", fontsize=20, y=1.02)
plt.show()
```

# 🔢 Activation Functions in Deep Learning

| Activation Function | Formula (Concept)                 | When to Use                                                                 | Notes                                                                      |
|---------------------|-----------------------------------|------------------------------------------------------------------------------|----------------------------------------------------------------------------|
| **Linear**          | `f(x) = x`                        | Output layer for regression tasks                                            | No non-linearity; not useful in hidden layers                             |
| **Sigmoid**         | `1 / (1 + e^-x)`                  | Binary classification output layer                                           | Can cause vanishing gradients; use with care                              |
| **Tanh**            | `(e^x - e^-x) / (e^x + e^-x)`      | Hidden layers for zero-centered outputs                                     | Better than sigmoid but still suffers from vanishing gradients            |
| **ReLU**            | `max(0, x)`                       | Default for hidden layers in CNNs, MLPs                                     | Fast and simple, but can cause "dying ReLU" problem                       |
| **Leaky ReLU**      | `x if x > 0 else αx`              | When ReLU units "die" (i.e., stuck at 0)                                     | Allows small gradient when x < 0                                          |
| **PReLU**           | `x if x > 0 else a*x (learned)`   | Similar to Leaky ReLU, but slope is learned                                 | Slightly better performance; added complexity                            |
| **ELU**             | `x if x > 0 else α*(e^x - 1)`     | Deep networks with potential for negative outputs                           | Helps with faster convergence                                             |
| **SELU**            | Scaled ELU                        | Self-normalizing neural networks                                            | Use with specific architectures; requires special initialization          |
| **Softplus**        | `log(1 + e^x)`                    | Smooth ReLU alternative                                                     | Never zero gradient, but slower                                            |
| **Softsign**        | `x / (1 + |x|)`                   | Like tanh but computationally lighter                                       | Slower convergence than tanh                                              |
| **Swish**           | `x * sigmoid(x)`                  | Image classification, NLP (advanced models)                                 | Smooth, non-monotonic, better accuracy than ReLU                          |
| **Mish**            | `x * tanh(softplus(x))`           | Vision & NLP tasks requiring smooth gradients                               | Similar to Swish, but more computationally intense                        |
| **GELU**            | `x * Φ(x)`                        | Transformer models (e.g., BERT, GPT), advanced NLP                          | Combines properties of ReLU & dropout                                     |
| **Hard Sigmoid**    | Piecewise sigmoid approx.         | Fast approximation in low-power hardware                                    | Less accurate, used in mobile/embedded settings                           |
| **Exponential**     | `e^x`                             | Experimental use; few practical applications                                | Rarely used in practice                                                   |
| **Bent Identity**   | `(sqrt(x^2 + 1) - 1)/2 + x`       | Hybrid of linear and non-linear behavior                                    | Smooth alternative to ReLU/tanh                                           |
| **Arctangent**      | `arctan(x)`                       | Older models or function approximation tasks                                | Slower convergence, bounded output                                        |
| **Sinc**            | `sin(x) / x`                      | Signal processing inspired applications                                     | Not commonly used in modern networks                                      |
| **Gaussian**        | `e^(-x^2)`                        | Radial basis function networks                                              | Good for localized response                                               |
| **Sign**            | `+1 if x>0, -1 if x<0`            | Binary thresholding or logic gates                                          | Non-differentiable; not trainable via gradient descent                    |



# ⚠️ Vanishing Gradient vs Exploding Gradient

These are two common problems that occur during backpropagation in deep neural networks.

---

## 🔻 Vanishing Gradient

### ❓ What is it?
When gradients become **very small** as they are backpropagated through many layers, causing **early layers to learn very slowly** or not at all.

### ⚙️ Why it Happens?
- Activation functions like `sigmoid` or `tanh` squash inputs to small ranges (e.g., 0 to 1), and their derivatives are also small (e.g., max derivative of sigmoid is 0.25).
- Multiplying many small gradients (in chain rule) leads to near-zero updates.

### 🧠 Symptoms
- Very slow or no training.
- Weights in early layers remain unchanged.
- Accuracy plateaus early.

### ✅ Solutions
- Use **ReLU** or its variants (`LeakyReLU`, `PReLU`, etc.).
- Use **Batch Normalization**.
- Use **residual connections** (ResNet).
- Use better weight initializations (e.g., **He initialization**).

---

## 🔺 Exploding Gradient

### ❓ What is it?
When gradients grow **exponentially large**, causing **unstable weight updates** and often resulting in `NaN` losses or model divergence.

### ⚙️ Why it Happens?
- Deep networks with large weights.
- Poor weight initialization.
- Repeated multiplication of large derivatives.
